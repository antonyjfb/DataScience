```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grafo_Erdos_Renyi(N,p,random_state=1):
    if random_state==0:
        np.random.seed(random_state)
    grafo=np.zeros((N,N))

    for i in range(N):
        for j in range(i+1,N):
            if np.random.random()<p:
                grafo[i,j]=1
                grafo[j,i]=1
    return grafo
```

```python
G=grafo_Erdos_Renyi(100,0.2)
```

```python
#Calculen el grado del grafo
grados=np.sum(G,axis=0)
```

```python
def buscar_componentes(grafo):
    N=len(grafo)
    visitados=[]
    componentes=[]
    for i in range(N):
        visitados.append(False)
    for i in range(N):
        if visitados[i]==False:
            componente=[]
            componentes.append(buscar_en_vecinos(componente,i,visitados,grafo))
    return componentes


def buscar_en_vecinos(componente,i,visitados,grafo):
    visitados[i]=True
    componente.append(i)
    vecinos=np.where(grafo[i,:]>0)[0]
    for j in vecinos:
        if visitados[j]==False:
            componente=buscar_en_vecinos(componente,j,visitados,grafo)
    
    return componente
    
```

```python
def extraer_componente_gigante(grafo):
    comps=buscar_componentes(grafo)
    comp_gig=comps[0]
    N_g=len(comp_gig)
    for comp in comps:
        if len(comp)>N_g:
            comp_gig=comp
            N_g=len(comp)
    comp_gig=np.sort(np.asarray(comp_gig))
    return comp_gig
```
