```python
from selenium import webdriver
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://www.vivanuncios.com.mx/s-venta-inmuebles/distrito-federal/page-2/v1c1097l1008p2'
```

```python
def get_url(num):
    url='https://www.vivanuncios.com.mx/s-venta-inmuebles'
    url+='/distrito-federal/page-{}/v1c1097l1008p{}'
    return url.format(num,num)
    
```

```python
driver = webdriver.Chrome()
driver.get(get_url(10))
```

```python
sopa=BeautifulSoup(driver.page_source, 'html.parser')
sopa
```

```python
res=sopa.find_all('div',{'class':'REAdTileV2'})
len(res)
```

```python
tarjeta=res[0]
tarjeta
```
